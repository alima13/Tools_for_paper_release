```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the accuracy data from the Excel file
df = pd.read_excel('1.xlsx')  # Replace 'accuracy_data.xlsx' with the actual file path

# Plot the comparison of accuracy for the models
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:  # Assuming the first column contains the iteration numbers
    plt.plot(df.iloc[:, 0], df[column], label=column)

plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Comparison of Accuracy for Different Attacks over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
